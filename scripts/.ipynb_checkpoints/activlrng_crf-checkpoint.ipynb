This code allows us to a CRF machine learning model to do morphological sequence labeling. With each model, we take the same three basic steps.

1) Train the model.  

2) Test the model

3) Predict tags over "unannotated" sequences. 

4) Sort predictions based on the model's "confidence" in its predictions. That is, the probabilities that the POS tagger computed for that particular sequence of labels. A higher average probablity == higher confidence in the predicted sequence of POS tags. 

5) Select N most uncertain predictions. Get all gold standard annotations for these words (multiple if more than one parse). Add these to training data.

6) Rinse and repeat

### Helper Functions
Here are some functions that will be handy later. The `cm` function builds and display a confusion matrix so we can evaluate where our tagger is making wrong predictions, after we test a POS tagger. The `sortUncertainty` function sorts our previously unannotated sentence based on how "confident" the model was at it predictions of each sentence's POS tags.

```python
def prepare(language, trainsize):
    '''Prepare data from formatted files'''
    infile = language+trainsize+'.input'
    outfile = language+trainsize+'.output'
    # read from files
    intrain = [line for line in DATA_FOLDERPATH+'train.'+infile]
    inselect = [line for line in DATA_FOLDERPATH+'select.'+infile]
    
    outtrain = [line for line in DATA_FOLDERPATH+'train.'+outfile]
    outselect = [line for line in DATA_FOLDERPATH+'select.'+outfile]
    
    intest = [line for line in DATA_FOLDERPATH+'test.'+language+'.input']
    outtest = [line for line in DATA_FOLDERPATH+'test.'+language+'.output']
        
    return intrain, outtrain, intest, outtest, inselect, outselect
```

```python
def cm(gold_tags, predicted_tags, taglist, modelname):
    # confusion matrix
    alpha_taglist = sorted(set(taglist))
    confusion_matrix = metrics.confusion_matrix(gold_tags,predicted_tags,labels=alpha_taglist,normalize="true")
    disp = metrics.ConfusionMatrixDisplay(confusion_matrix,display_labels=alpha_taglist)

    plt.rcParams["figure.figsize"] = (10,10)
    plt.rcParams.update({'font.size': 12})
    disp.plot(colorbar=False)
    plt.show() # display below
    plt.savefig(LOG_FOLDERPATH+ITERATION+modelname+'_matrix.png')    #save as file
    
    
    matrixstring = '{0:5s}'.format(' ') + '\t'.join(['{0:^4s}'.format(tag) for tag in alpha_taglist]) + '\n'
    for i,row in enumerate(confusion_matrix):
        cols = '\t'.join(['{:.2f}'.format(round(col,2)) for col in row])
        matrixstring+='{0:6s}'.format(alpha_taglist[i]) + cols + '\n'
    
    return matrixstring

    
def logResults(classreport, confusion, modelname):
    with open(LOG_FOLDERPATH+ITERATION+modelname+'.results', 'w') as R:
        time = str(datetime.datetime.now())
        report = '\n\nClassification Report\n\n{}\n\nConfusion Matrix\n\n{}\n'.format(classreport, confusion)
        R.write(time + report)
        
    
def sortUncertainty(w_t_predictions, seq_confidences_scores, modelname):
    '''auto-annotated sequences sorted by decreasing "confidence" scores,
    i.e., lower marginal probability == less confidence. 
    Writes to File.
    '''
    with_confidence = list(zip(w_t_predictions, seq_confidences_scores))
    with_confidence.sort(key = lambda x: x[1])
    sorted_sequences = [z[0] for z in with_confidence]
    predictFile(sorted_sequences, modelname)
```

## Model "CRF" - Conditional Random Fields (CRF) with Generic Features

A CRF is a sequence modeling algorithm useful for identifying entities or patterns in text, such as POS tags. This model assumes that features are dependent on each other, and it is able to consider future observations while learning a pattern. 

Machine learning models take into account previous data. We must feed into the CRF features from the data. The algorithm will create a predictive, probablistic model from these features. The word2features function below extracts features that characterize the data point, such as the sequence of previous words.

#This model has several good features for POS tagging. The original were taken from: https://towardsdatascience.com/pos-tagging-using-crfs-ea430c5fb78b#1c6a. The more English specific features were removed

```python
def extractWordFeatures(word, biolabels):
    '''Takes a line as list of [[word BIO-label]l]w
    Returns list of features for each letter.'''
    
    wfeaturelist = []
    
    # Define some features
    lettershape = 'lettershape='
    letterLowerCase = 'letterLowercase='
    fullword = 'fullword='
    wordshapelower = 'wordshaperlower='
    wordlength = 'wordlength='
    letterpositionend = 'letterpositionend='
    letterpositionbegin = 'letterpositionbegin='
    letterwindow = 4
    # optional
    postag = 'postag='
    
    letterindex = 0
    try:
        word = wordletters[0][0]
    except IndexError:
        print('NO WORD?:', wordletters)
    wordlen = len(word)
    for letterlist in wordletters:
        # initialize word/letter features
        wfeatures = ['bias',
                    lettershape + letterlist[1],
                    letterLowerCase + letterlist[1].lower(),
                    fullword + word,
                    wordshapelower + word.lower(),
                    wordlength + str(wordlen),
                    letterpositionend + str(letterindex - wordlen),
                    letterpositionbegin + str(letterindex+1)
                   ]
        if POS:
            wfeatures.append(postag + letterlist[2])
        # previous letters
        if letterindex == 0: # if first letter of word
            wfeatures.append('BOW')
        else:
            for i in range(1, letterwindow+1):
                if letterindex >= i:
                    wfeatures.append('prev' + str(i) + 'letters=' + 
                                    word[letterindex-i:letterindex].lower() + '>')
        # next letters
        if letterindex == wordlen-1: # if last letter of word
            wfeatures.append('EOW')
        else:
            for i in range(1, letterwindow+1):
                if letterindex + i < wordlen:
                    wfeatures.append('nxt' + str(i) + 'letters=<' + 
                                    word[letterindex+1:letterindex+1+i].lower())

        wfeaturelist.append([f.encode('utf-8') for f in wfeatures])
        letterindex += 1
        
    return wfeaturelist


def extractWordLabels(wordline):
    '''Takes a line as [[[word, letter, POS, BIO-label]l...]m...]w
    Returns list of bio-labels.'''
    
    wlabels = []
    for letterlist in wordline:
        wlabels.append(letterlist[-1].encode('utf-8'))
    return wlabels


def extractWords(wordline):
    '''Takes a line as [[[word, letter, POS, BIO-label]l...]m...]w
    Returns space delimited letters.'''
    
    return ' '.join(wordline[0])
```

```python
def featurize(data):
    return [extractWordFeatures(word) for word in data]


def getLabels(data):
    return [extractWordLabels(word) for word in data]


# words and labels, feed words to feature extractor
def word2features(sent):
    # remove spaces between letters of words for morphology
    return [word2features(''.join(sent.split()), i) for i in range(len(sent))]

def word2labels(sent):
    return [word[1] for word in sent]

def word2tokens(sent):
    return [word[0] for word in sent]
```

```python
def mainCRF(language, trainsize, selectsize):
    
    Xtrain, ytrain, Xtest, ytest, Xselect, yselect = prepare(language, trainsize)
    
    # get BIO labels
    
    
    # extracting features from all the sentences
    train_ftrs = [word2features(s) for s in Xtrain]
    train_tags = [word2labels(s) for s in ytrain]

    test_ftrs = [word2features(s) for s in Xtest]
    test_tags = [word2labels(s) for s in ytest]
    
    unannotated_ftrs = [word2features(s) for s in Xselect]
    unannotated_words = [word2labels(s) for s in yselect]
    
    # training
    crf = sklearn_crfsuite.CRF(
        algorithm = 'lbfgs',
        c1 = 0.25,
        c2 = 0.3,
        max_iterations = 100,
        all_possible_transitions=True)
    
    crf.fit(train_ftrs, train_tags)
    
    # testing
    test_output = crf.predict(test_ftrs)
    # test reports
    test_tags = list(itertools.chain.from_iterable(test_tags))
    test_output = list(itertools.chain.from_iterable(test_output))
    scores = metrics.classification_report(test_tags, test_output, zero_division=0.0)
    matrix = cm(test_tags, test_output, tags, modelname)
    logResults(scores, matrix, modelname)
    
    # predict
    predicted_labels = crf.predict(unannotated_ftrs)
    predicted_sequences = [list(zip(unannotated_words[i],predicted_labels[i])) for i in range(len(predicted_labels))]
    all_probs = crf.predict_marginals(unannotated_ftrs)
    
    confidences = []
    for s,sent in enumerate(predicted_sequences):
        confidences.append(sum([all_probs[s][i][pair[1]] for i,pair in enumerate(sent)])/len(sent))
    
    sortUncertainty(predicted_sequences, confidences, modelname)
```
